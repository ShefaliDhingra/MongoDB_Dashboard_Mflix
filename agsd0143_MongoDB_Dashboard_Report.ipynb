## **Description of Data**
The MFlix dataset is a sample movie database provided for MongoDB applications. It contains structured and semi-structured data about movies, including:  
- Titles, genres, and release years  
- Ratings (IMDB, Metacritic, viewer ratings)  
- Content classifications  
- Languages and production locations  
- Directors and their works  
- Awards and nominations  

The dataset serves as a foundation for building an analytical dashboard that provides insights into movie trends over time.  

---

## **Dataset Attributes Description**
The MFlix dataset consists of multiple fields, including:  
- **_id**: Unique identifier for each movie record.  
- **title**: Name of the movie.  
- **year**: Release year of the movie.  
- **genres**: List of genres associated with the movie.  
- **runtime**: Duration of the movie in minutes.  
- **rated**: Content rating (e.g., PG, R, G, etc.).  
- **cast**: List of actors in the movie.  
- **directors**: List of directors associated with the movie.  
- **languages**: Languages used in the movie.  
- **countries**: Countries where the movie was produced.  
- **imdb**: IMDb-related information (rating, votes, ID).  
- **metacritic**: Metacritic rating of the movie.  
- **awards**: Award nominations and wins.  
- **comments**: Viewer comments and engagement metrics.
